# Task 3.2: Effect of increasing bit-error probability on narrow-band speech with and without a (3,1) repetition FEC scheme

### Part 1: 13 versions of speech

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.io import wavfile
from comp28512_utils import Audio, insert_bit_errors, get_pesq_scores, numpy_array_to_bit_array, bit_array_to_numpy_array, get_pesq_scores

(Fs, speech) = wavfile.read("NarrobandSpeech8k.wav")

speechToBits = numpy_array_to_bit_array(speech)

x = np.zeros(14)

y = np.zeros(14)

probabilityOfBits = 0.00001
for i in range(0,14):
    bitsWithErrors = insert_bit_errors(speechToBits, probabilityOfBits)
    
    bitsToSpeech = bit_array_to_numpy_array(bitsWithErrors, dtype=speech.dtype)

    Audio(bitsToSpeech, rate = Fs)
    probabilityOfBits += (0.01 - 0.00001) / 13 # 0.00001 -> 0.01
    y[i] = probabilityOfBits
    wavfile.write("bitErrorSpeech.wav",Fs,bitsToSpeech)
    ! linux_pesqmain +8000 NarrobandSpeech8k.wav bitErrorSpeech.wav > /dev/null
    pesq_results = get_pesq_scores()
    score = pesq_results["NarrobandSpeech8k.wav"] ["bitErrorSpeech.wav"] 
    print "Score of file with probability = ", probabilityOfBits," :", score
    x[i] = score

```

### Part 2: PESQ Graph

```python
plt.semilogx(x,y) 
plt.grid(True)
plt.title('PESQ SCORES')
plt.xlabel('Bit-error probability')
plt.ylabel('PESQ score')
plt.show()
```

# Task 3.3 Apply ARQ to a text message

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.io import wavfile
from comp28512_utils import Audio, insert_bit_errors, get_pesq_scores, numpy_array_to_bit_array, bit_array_to_numpy_array

def CRC8(xa):
    g8x = [1, 0, 0, 0, 0, 0, 1, 1, 1]
    xae = [xa, 0, 0, 0, 0, 0, 0, 0, 0]
    xsa = xae[1:9]
    for i in (1, len(xa)):
        if xsa[1] == g8x[1]:
            xsa = xsa ^ g8x
        xsa[1:8] = xsa[2:9]
        if i < len(xa):
            xsa[9] = xae[i+9]
    
    return xsa[1:8]
s = "Behold the sea itself, and on its limitless heaving breast, the ships."
CRC8(numpy_array_to_bit_array(s)[0:8])
```
